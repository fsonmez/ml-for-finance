```python
from pandas_datareader import data
import pandas as pd
import numpy as np
```

```python
start_date = '2014-01-01'
end_date = '2018-01-01'
goog_data = data.DataReader('goog', 'yahoo', start_date, end_date)
```

```python
goog_data.head(10)
```

We create a new dataframe carrying the signal we will processing. We will use the adjusted close price in the above dataframe

```python
goog_data_signal = pd.DataFrame(index=goog_data.index)
goog_data_signal['price'] = goog_data['Adj Close']
goog_data_signal['daily_difference'] = goog_data_signal.price.diff()
```

```python
goog_data_signal['signal'] = 0.0
goog_data_signal.signal = np.where(goog_data_signal.daily_difference > 0, 1, 0)
goog_data_signal['positions'] = goog_data_signal.signal.diff()
```

```python
goog_data_signal.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(111, ylabel ='Google price in $')
goog_data_signal.price.plot(ax=ax1, color='r',lw=2)
# plot up arrows
ax1.plot(goog_data_signal.loc[goog_data_signal.positions == 1].index,
        goog_data_signal.price[goog_data_signal.positions == 1],
        '^',
        markersize=5,
        color='m'
        )
#plot down arrows
ax1.plot(goog_data_signal.loc[goog_data_signal.positions == -1].index, 
        goog_data_signal.price[goog_data_signal.positions == -1],
        'v',
        markersize=5,
        color='k'
        )
```

Now we move to backtasting the strategy. We will assume that the porfolio consists only of one stock (GOOG) and that we have $1000 initial capital.


```python
initial_capital = float(1000)
positions = pd.DataFrame(index=goog_data.index).fillna(0.0)
portfolio = pd.DataFrame(index=goog_data.index).fillna(0.0)
```

```python
positions['GOOG'] = goog_data_signal.signal
portfolio['positions'] = positions.multiply(goog_data_signal.price, axis=0)
positions.head()
```

```python
portfolio['cash'] = initial_capital - (positions.diff().multiply(goog_data_signal.price, axis=0)).cumsum()
```

```python
portfolio['total'] = portfolio['positions'] + portfolio['cash']
```

```python
plt.style.use('dark_background')
fig2 = plt.figure(figsize=(12,8))
ax2 = fig2.add_subplot(111)
portfolio.positions.plot(ax=ax2, alpha = 0.2)
portfolio.cash.plot(ax=ax2, alpha=0.5)
portfolio.total.plot(ax=ax2)
```
